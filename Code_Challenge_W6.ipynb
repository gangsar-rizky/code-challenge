```python
# delete null value
dropped_df = df.copy()
dropped_df = dropped_df[dropped_df['Manufacturer'].notna()]
#dropped_df = dropped_df[dropped_df['Make'].notna()]
#dropped_df.dropna(axis=0, inplace=True)
dropped_df
```

```python
replaced_value_df = dropped_df.copy()
replaced_value_df['Price'].fillna(
    dropped_df['Price'].mean(), axis=0, inplace=True)
replaced_value_df.head()
```

```python
# remove duplicates
replaced_value_df[dropped_df['Model'].duplicated()]
# no duplicates found in column 'Model'
```

```python
# OUTLIER
replaced_value_df.info()
```

```python
df_USA = replaced_value_df[replaced_value_df['Origin']== 'USA'].copy()
df_USA['Price'] = round(df_USA.Price, 1)
df_USA.head()
```

```python
df_non = replaced_value_df[replaced_value_df['Origin']== 'non-USA'].copy()
df_non['Price'] = round(df_non.Price,1)
df_non.head()
```

```python
df_USA.mean()
```

```python
df_USA.boxplot(column=['Price'], fontsize=10,
               rot=0,grid=False,figsize=(10,10),vert=False)
```

```python
Q1_USA = df_USA['Price'].quantile(0.25)
Q3_USA = df_USA['Price'].quantile(0.75)

IQR_USA = Q3_USA - Q1_USA
boxplot_min_USA = Q1_USA - 1.5 * IQR_USA
boxplot_max_USA = Q3_USA + 1.5 * IQR_USA
print('Q1_USA:\n',Q1_USA)
print('\nQ3_USA:\n', Q3_USA)
print('\nIQR_USA:\n', IQR_USA)
print('\nMin:\n', boxplot_min_USA)
print('\nMax:\n', boxplot_max_USA)
```

```python
filter_min_USA = df_USA['Price']<boxplot_min_USA
filter_max_USA = df_USA['Price']>boxplot_max_USA

df_USA_non_outlier = df_USA[~
                  (filter_min_USA|filter_max_USA)          
                            ]
df_USA_non_outlier
```

```python
data_us = df_USA_non_outlier
data_us
data_us.corr()
```
